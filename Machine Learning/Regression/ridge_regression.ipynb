```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

```python
# Step 1: Generate synthetic data
np.random.seed(420)  # For reproducibility
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relationship with noise
```

```python
# Step 2: Add intermittent noise
noise_indices = np.random.choice(100, 10, replace=False)
y[noise_indices] += 10 * np.random.randn(10, 1)
```

```python
# Step 3: Apply ridge regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X, y)
y_pred = ridge_reg.predict(X)
```

```python
# Step 4: Visualize the results
plt.figure(figsize=(12, 6))

# Plot original noisy data
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Noisy data')
plt.plot(X, 4 + 3 * X, color='red', label='True line')
plt.title('Original Noisy Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Plot ridge regression results
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Noisy data')
plt.plot(X, y_pred, color='green', label='Ridge regression line')
plt.title('Ridge Regression Results')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
